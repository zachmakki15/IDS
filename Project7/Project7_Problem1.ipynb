##Disease_1 Column

A new column diease_1 will be added to the dataset. This column will be ordinal with potential values being 0 and 1. 1 when the class # in the target attribute is #1 and 0 for anything else.

```python
disease_1 = []
for row in dataFrame.index:
  if (dataFrame['class'][row] == 1):
    disease_1.append(1)
  else:
    disease_1.append(0)

len(disease_1)
```

```python
#Add newly created list as new column in dataframe
dataFrame['disease_1'] = disease_1
```

##Models

###Linear Regression Model

A linear regression will be created to see if disease_1 can be predicted based on the attribute age_imputed (age without null values)

```python
#Scatter plot of disease_1 vs age
sns.pairplot(dataFrame, x_vars=['age_imputed'], y_vars=['disease_1'], height=4, aspect=1, kind='scatter')
```

```python
#Linear regression for disease_1 and age
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
lm = linear_model.LinearRegression()

#Create x and y
x = dataFrame['age_imputed'].values.reshape(-1,1)
y = dataFrame['disease_1'].values.reshape(-1,1)

#Use 3-cross-folds validation
r2_scores = cross_val_score(lm, x, y, scoring='r2', cv=3)
#Display r2 scores
print(r2_scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (r2_scores.mean(), r2_scores.std() * 2))
```

Since one of the attributes in this regression is categoric it makes sense that the r2 numbers would be so small.

```python
#Fit model
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.7,test_size=0.3,random_state=42, stratify=y)
lm.fit(x_train, y_train)

#Get the intercept and slope
t0, t1 = lm.intercept_[0], lm.coef_[0][0]
t0, t1

#Plot the linear regression just created
dataFrame.plot(kind='scatter', x='age', y='disease_1', figsize=(8, 6))
plt.axis([0, 80, -.5, 1.5])
plt.plot(x, t0 + t1*x, "r")
plt.show()
```

```python
#Get MSE and R2 score for non-cross fold test
#This is done because the sklearn cross folds method does not compute mse
from sklearn.metrics import r2_score, mean_squared_error

#Next train the model using the test data
y_test_predictions = lm.predict(x_test)

#Print the scores
print('R2 score:',            r2_score(y_test, y_test_predictions))
print('Mean squared error:', mean_squared_error(y_test, y_test_predictions))
```

The MSE score matches the r2 score as its value indicates that the regression done above is not a good predictor of disease_1. 

###NB Classifier Model

The best 5 attributes out of all the crosstabs shown above will be selected and used as the x to predict the disease_1 attribute via a Naive Bayes model. The 5 attributes that will be selected as the x for this model are fibrosis_of_the_papillary_dermis, perifollicular_parakeratosis, thinning_of_the_suprapapillary_epidermis, clubbing_of_the_rete_ridges, and scalp_involvement.

```python
#Import libs needed for NB model
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

#Create new x and y (x=the top ranking 5 attributes) (y=disease_1)
x = dataFrame[['fibrosis_of_the_papillary_dermis', 'perifollicular_parakeratosis', 'thinning_of_the_suprapapillary_epidermis', 
               'clubbing_of_the_rete_ridges', 'scalp_involvement']]
y = dataFrame['disease_1']

#3-cross-folds validation test
NBmodel = GaussianNB()
nb_scores = cross_val_score(NBmodel, x, y, cv=3)

#Show scores
print(nb_scores)

#Show score mean and std
print("Accuracy: %0.2f (+/- %0.2f)" % (nb_scores.mean(), nb_scores.std() * 2))
```

From the results of the cross validation we can see that the NB model did not perform very well on this dataset. Reasons for this could be that I did not select the correct attributes from the cross-tabulation section, or the attributes selected do not lend themselves well to a NB model. 

```python
#Show NB model confusion matrix
#Split data for training/testing
from sklearn.metrics import confusion_matrix
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=42, stratify=y)

#Fit model
NBmodel.fit(x_train, y_train)

#Get model predictions
y_pred = NBmodel.predict(x_test)

#Create confusion matrix
labels = [0,1]
confusionMatrix = confusion_matrix(y_test, y_pred, labels=labels)
print('Confusion matrix:')
print(confusionMatrix)
pd.DataFrame(confusionMatrix, index=labels, columns=labels)
sns.heatmap(confusionMatrix, annot=True, linewidths=.5, square = True, cmap="BuPu", fmt='0.4g')
plt.ylabel('Actual')
plt.xlabel('Predicted')
```

The confusion matrix shown here reinforces the results of the cross validation show above.

###DT Model

A decision tree model will be created for the entire dataset with the original target attribute (class) as the y.

```python
#Import libs needed for DT model
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

#Create new x and y (x=full dataset-class) (y=class)
x = dataFrame.drop(columns=['age', 'disease_1', 'class'])
y = dataFrame['class']

#Test with 3-cross folds validation
DTmodel = DecisionTreeClassifier(criterion = 'gini', random_state = 42)
dt_scores = cross_val_score(DTmodel, x, y, cv=3)

#Show scores
print(dt_scores)

#Show scores mean and std
print("Accuracy: %0.2f (+/- %0.2f)" % (dt_scores.mean(), dt_scores.std() * 2))
```

The results of the cross validation done above show the DT model was able to accuratly predict class using the entire dataset as x.

```python
#Display DT model tree and create confusion matrix
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.7,test_size=0.3,random_state=42, stratify=y)

#Fit model
DTmodel.fit(x_train, y_train)

#Display tree
from matplotlib.pyplot import figure
figure(figsize=(20, 15))
tree.plot_tree(DTmodel, feature_names=x.columns.values.tolist(), filled=True)
text_representation = tree.export_text(DTmodel, feature_names=x.columns.values.tolist())
print(text_representation)
```

```python
#Show DT confusion matrix
y_pred = DTmodel.predict(x_test)

#Create confusion matrix
labels = [1,2,3,4,5,6]
confusionMatrix = confusion_matrix(y_test, y_pred, labels=labels)
print('Confusion matrix:')
print(confusionMatrix)
pd.DataFrame(confusionMatrix, index=labels, columns=labels)
sns.heatmap(confusionMatrix, annot=True, linewidths=.5, square = True, cmap="BuPu", fmt='0.4g')
plt.ylabel('Actual')
plt.xlabel('Predicted')
```

The confusion matrix above confirms the results from the cross validation, showing a well performing model.

###RF Model

Since the DT model performed well on this dataset, one can reason that maybe a RF model could be even better. That will be tested below.

```python
#Import lib needed for RF model
from sklearn.ensemble import RandomForestClassifier

#Use 3-cross-folds validation to test the model
RFmodel = RandomForestClassifier(n_estimators=500)
rf_scores = cross_val_score(RFmodel, x, y, cv=3)

#Show scores
print(rf_scores)

#Show score mean
print("Accuracy: %0.2f (+/- %0.2f)" % (rf_scores.mean(), rf_scores.std() * 2))
```

As predicited above the RF model does perform slightly better then the DT model on the entire dataset with class as the target (y).

```python
#Show info about one of the trees in a RF
RFmodel.fit(x_train,y_train)

#Extract single tree
oneTreeHill = RFmodel.estimators_[42]

#Display single tree
figure(figsize=(20, 15))
tree.plot_tree(oneTreeHill, feature_names=x_train.columns.values.tolist(), filled=True)
text_representation = tree.export_text(oneTreeHill, feature_names=x_train.columns.values.tolist())
print(text_representation)
```

```python
#Show model confusion matrix
y_pred = RFmodel.predict(x_test)
confusionMatrix = confusion_matrix(y_test, y_pred, labels=labels)
print('Confusion matrix:')
print(confusionMatrix)
pd.DataFrame(confusionMatrix, index=labels, columns=labels)
sns.heatmap(confusionMatrix, annot=True, linewidths=.5, square = True, cmap="BuPu", fmt='0.4g')
plt.ylabel('Actual')
plt.xlabel('Predicted')
```

This confusion matrix confirms the results of the cross validation done above and the slight improvement in the RF model becomes apparent when compared to the confusion matrix from the DT model.

```python
#Plot feature importance
features = list(x_train.columns)
importances = RFmodel.feature_importances_
indices = np.argsort(importances)

figure(figsize=(20, 15))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='lime', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

Based on the results of the feature importance plot the top 3 most important features in the dataset for the RF model are clubbing_of_the_rete_ridges, fibrosis_of_the_papillary_dermis, and thinning_of_the_suprapapillary_epidermis. Interestly these three attributes were all selected as part of the top 5 performing attributes in the NB model section of this report.

```python
#Compare models
labels = ['LR', 'NB', 'DT', 'RF']
model_scores = [r2_scores.mean(), nb_scores.mean(), dt_scores.mean(), rf_scores.mean()]

x = np.arange(len(labels))
width = 0.15
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, model_scores, width, label='Model Scores', color='purple')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Model Accuracy')
ax.set_xticks(x)
ax.set_ylim(0,1)
ax.set_xticklabels(labels)
ax.legend(bbox_to_anchor = (1.4, 1.0), loc='upper right')
plt.title('Model Comparison')
plt.show()
```
